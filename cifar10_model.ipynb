```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, datasets
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# 1. а№Ђаё•аёЈаёµаёўаёЎаё‚а№‰аё­аёЎаё№аёҐ
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
```

```python
# 2. Data Augmentation (аёЄаёЈа№‰аёІаё‡аё аёІаёћа№Ђаёћаёґа№€аёЎа№ЃаёљаёљаёЄаёёа№€аёЎ)
datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
)
datagen.fit(x_train)
```

```python
# 3. аёЄаёЈа№‰аёІаё‡ Model аё—аёµа№€аёҐаё¶аёЃаё‚аё¶а№‰аё™аёћаёЈа№‰аё­аёЎ Batch Normalization а№ЃаёҐаё° Dropout
model = keras.Sequential([
    # Block 1
    layers.Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(32, 32, 3)),
    layers.BatchNormalization(),
    layers.Conv2D(32, (3, 3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.2),

    # Block 2
    layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.3),

    # Block 3
    layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.4),

    # Fully Connected
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])
```

```python
# 4. Compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# 5. Train а№‚аё”аёўа№ѓаёЉа№‰ Data Augmentation
# а№Ђаёћаёґа№€аёЎ epochs а№Ђаё›а№‡аё™ 50-100 а№Ђаёћаё·а№€аё­аё„аё§аёІаёЎа№ЃаёЎа№€аё™аёўаёіаёЄаё№аё‡аёЄаёёаё”
model.fit(datagen.flow(x_train, y_train, batch_size=64),
          epochs=50,
          validation_data=(x_test, y_test))

model.save('cifar10_improved_model.h5')
```
